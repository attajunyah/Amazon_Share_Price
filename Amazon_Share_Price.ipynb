```python
%matplotlib inline
import pandas as pd 
import mplfinance as mpf
```

```python
file = 'AMZN.csv'
data = pd.read_csv(file)
```

```python
data.Date = pd.to_datetime(data.Date)
data.info()
```

```python
data = data.set_index('Date')
```

```python
data
```

```python
mpf.plot(data)
```

```python
mpf.plot(data,type='line', volume=True)
```

```python
#How do we plot a single month's data?
mpf.plot(data['2021-03'], volume=True)
```

```python
mpf.plot(data['2021-03': '2021-07'], type='candle', mav=(20), volume=True, tight_layout=True)
```

```python
mpf.plot(data['2021-03': '2021-07'], figratio=(20,12), 
         type='candle', 
         title = 'Amazon Price 2021/2022',
         mav=(20), volume=True, 
         tight_layout=True)
```

```python
mpf.plot(data['2021-03': '2021-07'], figratio=(20,12), 
         type='candle', 
         title = 'Amazon Price 2021/2022',
         mav=(20), volume=True, 
         tight_layout=True, style='yahoo')
```
